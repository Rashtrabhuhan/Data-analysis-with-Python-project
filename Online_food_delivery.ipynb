# ***Visualization***

```python
ax = sns.countplot(x = "Gender", data = df, color="#7D8D86")
sns.set_style("darkgrid")
ax.bar_label(ax.containers[0])
plt.title("Compare of Gender")
plt.show()
```

* From above graph we can see that most of the order buyers are male and even the food ordering power of male are greater then female.

```python
ax = sns.countplot(x = "Medium (P2)", data = df,color="#7E6363")
sns.set_style("darkgrid")
ax.bar_label(ax.containers[0])
plt.title("Purchasing Method")
plt.show()
```

^ From above graph we can see that most of the order booked on direct phone call

```python
gb1 = df.groupby("Educational Qualifications").agg({"Educational Qualifications" : "count"})
plt.pie(x = "Educational Qualifications", labels = gb1.index, data = gb1, autopct="%0.1f%%", radius = 1.5)
plt.show()
```

```python
ax1 = sns.countplot(x = "Educational Qualifications",data = df,color="#635985")
sns.set_style("darkgrid")
ax1.bar_label(ax1.containers[0])
plt.title("qualification Besed Order Quantity")
plt.show()
```

* From above graph we can see that most of the foof order by the Graduate and Post Graduate

```python
plt.figure(figsize= (9,5))
gb2 = df.groupby("Monthly Income").agg({"Monthly Income" : "count"})
sns.lineplot(x =gb2.index, y = gb2["Monthly Income"],data = gb2)
sns.set_style("darkgrid")
plt.title("Order besed on Income")
plt.show()
```

```python
sns.violinplot(x = "Educational Qualifications", y = "Monthly Income", data = df)
sns.set_style("darkgrid")
plt.show()
```

* From the above graphs we can see that the most of the people food ordering have no income

```python
gb3 = df.groupby("Occupation").agg({"Occupation" : "count"})
ax3 = sns.barplot(x = gb3.index,y = "Occupation", data = gb3, color="#373A40")
sns.set_style("darkgrid")
ax3.bar_label(ax3.containers[0])
plt.title("Order Quantity Besed on Occupation")
plt.show()
```

* From above graph we can see that most of the buyers are students accoding to Occupation

```python
ax4 = sns.countplot(x = "Marital Status", data = df, hue = "Gender",color="#A6B28B")
for bar in ax4.containers:
  ax4.bar_label(bar)
sns.set_style("darkgrid")
plt.show()
```

* From above graph we can see that most of the buyers are sinsle (male)

```python
gbx = df.groupby("Meal(P2)").agg({"Meal(P2)" : "count"})
ex = [0.0,0.2,0.1]
plt.pie(x = "Meal(P2)", labels =gbx.index, data = gbx, radius=1.5, autopct="%0.2f%%",explode= ex)
plt.show()
```

* From above graph we can see that the 80% order in Dinner time. and very less quantity of order in Lunch time

```python
gbxx = df.groupby("Family size").agg({"Family size" : "count"})
ax5 = sns.barplot(x = gbxx.index,y = "Family size", data = gbxx,color="#A35C7A")
sns.set_style("darkgrid")
ax5.bar_label(ax5.containers[0])
plt.show()
```

* Fron above graphs we can see that most of the buyers are of familysize of member 3 or 2

```python
plt.figure(figsize = (10,5))
ax6 = sns.countplot(x ="Maximum wait time", data = df,color="#239BA7")
ax6.bar_label(ax6.containers[0])
sns.set_style("darkgrid")
plt.xticks(rotation = 20)
plt.show()
```

* from above graph we can see that the wait time is more then the 45 mint then order quantity is very less
